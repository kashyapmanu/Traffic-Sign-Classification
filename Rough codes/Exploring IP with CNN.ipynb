```python
import cv2
import seaborn as sns
import matplotlib.pyplot as plt

#Reading the image
image = cv2.imread('goku.png')

#OpenCV loads in BGR format by default so we convert it into RGB
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show() #display the image in a cartesian plane

image.shape
```

```python
#Getting the value of the each pixel of the image
image[0][0] #White color
```

```python
#converting the image to grey color
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
originalImage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

print("The original image")
plt.imshow(originalImage)
plt.show()

print("The shape of the original image: ", originalImage.shape)

print("\n****************************************************\n")

print("The grayscale image")
plt.imshow(grayImage)
plt.show()

print("The shape of the grayscale image: ", grayImage.shape)
```

```python
#Converting the pixel values from teh greyscale image to a tabular form

import numpy as np
data = np.array(grayImage)
print("Before flattening: ",data.shape)
flattenedData = data.flatten()
print("After flattening:",flattenedData.shape)
flattenedData
```
